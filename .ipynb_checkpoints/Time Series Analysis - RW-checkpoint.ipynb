```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline
```

```python
# Reading Up To data
df_upto = pd.read_csv('Trend 2017-2019 - UpTo.csv')
df_upto.head(3)
```

```python
### Transforming original dataset into required format, with Date as DateTime and set to index

df_upto.Date = pd.to_datetime(df_upto.Date)
df_upto.set_index(df_upto.Date, inplace=True)
df_upto.drop(columns='Date', inplace=True)
```

```python
df_upto.head(3)
```

### Classical Decomposition (naive)

```python
result = sm.tsa.seasonal_decompose(df_upto,model='additive')
result.plot();
```

```python
plt.figure(figsize=(12,3))
plt.plot(result.observed);
plt.xlim(['2017-01-01', '2019-08-01'])
plt.title("Decomposition Plot - Observed");
```

```python
plt.figure(figsize=(12,3))
plt.plot(result.trend);
plt.xlim(['2017-01-01', '2019-08-01'])
plt.title("Decomposition Plot - Trend");
```

```python
plt.figure(figsize=(12,3))
plt.plot(result.seasonal);
plt.xlim(['2017-01-01', '2019-08-01'])
plt.title("Decomposition Plot - Seasonality");
```

```python
plt.figure(figsize=(12,3))
plt.plot(result.resid);
plt.xlim(['2017-01-01', '2019-08-01'])
plt.title("Decomposition Plot - Residuals");
```

# Model Training
Now that we have decomposed the Time Series, it will provide valuable insight into the most likely appropriate parameters to use with time series forecasting models.
Let's compare the performance of two of the most typical approaches: 1) ETS (AKA Holt-Winters); and 2) ARIMA (and its seasonal variations, SARIMAX).

## Pre-processing
Adjust the dataframe to work with `statsmodels` implementations, as well as setting the appropriate index as data-time, set target feature as double datatype (required by `statsmodels`), and split training/validation dataset:

```python
# Check index is set as datetime data type
assert type(df_upto.index) == pd.core.indexes.datetimes.DatetimeIndex
```

```python
df_upto.UpTo = df_upto.UpTo.astype('double')
```

```python
# Split Train/Validation. Last 3 months as validation

df_train = df_upto.iloc[:-3,:]
df_test = df_upto.iloc[-3:,:]
```

## 1) ETS (Holt-Winters)

```python
try:
    v = sm.version.version
except:
    v = sm.__version__
print(f'Statsmodels version has to be >= 0.10 | Current version is {v}')

# ! pip install statsmodels --upgrade

# Requires to restart kernel after installation
```

```python
# Necessary to use latest version of statsmodels (>= 0.10), otherwise throws error. 
from statsmodels.tsa.holtwinters import ExponentialSmoothing as ETS

model_ETS = ETS(df_train, trend='add', damped=True, seasonal_periods=12, seasonal='add')
```

```python
forecast = model_ETS.fit()
```

```python
yhat = forecast.forecast(3)
```

```python
plt.figure(figsize=(18,5))
plt.plot(df_test, ls='-', c='b', label='Validation Set')
plt.plot(yhat, ls=':', c='r', label='ETS prediction')
plt.plot(df_train, ls='-', c='k', label='Original Data')
plt.title('Up To: ETS forecast of last 6 months')
plt.legend()
plt.ylabel('Up To');
print(yhat)
```

```python
score_ETS = forecast.aic
print(f'ETS model AIC: {score_ETS:,.2f}')
```

Since Partial Auto Correlation plot drastically cuts off at lag 2, we identify it as the ideal amount of periods to be considered in the Auto Regression component.

```python
import itertools
```

```python
# Setting up space range to test and find optimum model
p=d=q= range(0,3)
pdq = list(itertools.product(p,d,q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p,d,q))]
```

```python
import warnings
warnings.filterwarnings("ignore", module='sm')
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model = sm.tsa.statespace.SARIMAX(df2.y,
                                         order=param,
                                         seasonal_order=param_seasonal,
                                         enforce_stationarity=False, enforce_invertibility=False)
            results = model.fit()
            print(f'ARIMA{param}x{param_seasonal}12 - AIC: {results.aic:.2f}')
            
        except: continue
```

```python
model_ARIMA = sm.tsa.statespace.SARIMAX(df2.y,
                                       order=(2,2,0),
                                       seasonal_order=(0,2,0,12),
                                       enforce_stationarity=False, enforce_invertibility=False)
results = model_ARIMA.fit()
results.summary()
```

```python
results.plot_diagnostics(figsize=(16,8));
```

```python
pred = results.get_prediction(start='2019-01-01', stop='2019-07-01', dynamic=False)

pred_ci = pred.conf_int()

plt.figure(figsize=(8,4))

plt.plot(df2.y, label='Observed');
plt.plot(pred.predicted_mean, alpha=0.7, label='One-stepp ahead Forecast');

plt.fill_between(pred_ci.index, pred_ci.iloc[:,0], pred_ci.iloc[:,1], color='k', alpha=0.1)
plt.ylabel('Up To')
plt.xlabel('Date')
plt.legend()
plt.tight_layout();
```

## 3) Prophet (Facebook)
Some extra steps are required to get this amazing package to work. 
1. Install pystan
2. Install fbprophet

These have been added in the `requirements` file and should be auto-installed when setting up the environment.

This package requires the dataframe to be structured in a very particular manner, having columns (not the datetime index) as:
+ Date component named as "ds"
+ Feature of interest named as "y"

```python
# Create a copy to prepare column names as required
df2 = df_upto.copy()
```

```python
# remove date as index
df2.reset_index(inplace=True)

# rename columns as required
df2.rename(columns={'Date':'ds', 'UpTo':'y'}, inplace=True)
df2.set_index(df2.ds, inplace=True)

df2.head(2)
```

```python
# Split training and validation data
df2_train = df2.loc[df2.index[:-3]]
df2_test = df2.loc[df2.index[-3:]]
```

```python
from fbprophet import Prophet
model = Prophet(n_changepoints=20)
model.fit(df2_train);
```

```python
# Create date time data to be forecasted
future_data = model.make_future_dataframe(3, freq='m')
```

```python
# Actually perform forecasting with the trained model
forecast_data = model.predict(future_data)
```

```python
# Plot predictions vs actual values
model.plot(forecast_data);
plt.title('Prophet Predictions')
plt.ylabel('Up To')
plt.xlabel('Date');
plt.plot(df_test, ls='-', c='b', label='Validation Set')
plt.legend();
```

```python
# Plot influence of Time Series components
model.plot_components(forecast_data);
```
